### **Analisis De Outliers**

```python
import plotly.express as px

numeric_features = [
    'Age', 'BMI', 'AlcoholConsumption', 'PhysicalActivity', 'DietQuality',
    'SleepQuality', 'SystolicBP', 'DiastolicBP', 'CholesterolTotal',
    'CholesterolLDL', 'CholesterolHDL', 'CholesterolTriglycerides',
    'MMSE', 'FunctionalAssessment', 'ADL'
]


fig = px.box(
    data[numeric_features].melt(var_name="Variable", value_name="Valor"),
    x="Variable",
    y="Valor"
)

fig.update_layout(height=700, width=1400)
fig.show()
```

**Conclusión**

**1. Rango de valores y dispersión:**

Variables como CholesterolTriglycerides, CholesterolTotal y SystolicBP presentan rangos amplios y variabilidad considerable.

Variables como PhysicalActivity, DietQuality y SleepQuality tienen rangos más pequeños y menos dispersión.

**2. Outliers visibles:**

Se observan posibles outliers en BMI, AlcoholConsumption, SystolicBP, DiastolicBP y CholesterolTriglycerides, ya que algunos valores se encuentran fuera de los bigotes del boxplot.

Esto indica que hay pacientes con valores atípicos que podrían afectar análisis estadísticos si no se manejan adecuadamente.

**3. Comparación de mediana y simetría:**

Variables como Age, BMI y CholesterolTotal muestran una mediana centrada y una dispersión moderada.

Variables como CholesterolTriglycerides parecen sesgadas hacia valores altos, con mediana más baja que la media, lo que indica asimetría positiva.

**4. Variables con menor variabilidad:**

PhysicalActivity, DietQuality, SleepQuality y FunctionalAssessment presentan menor dispersión, indicando que la mayoría de los pacientes tienen valores similares en estas variables.

### **Matriz Correlación**

---



```python
# Calcular matriz de correlación
corr = data.corr(numeric_only=True)

# Crear máscara para ocultar la mitad superior
mask = np.triu(np.ones_like(corr, dtype=bool))

# Aplicar máscara (reemplazar mitad superior con NaN)
corr_masked = corr.mask(mask)

# Heatmap con Plotly
fig = px.imshow(
    corr_masked,
    text_auto=True,
    color_continuous_scale='RdBu_r',
    aspect="auto",
    title="Matriz de correlación (solo mitad inferior)"
)

fig.update_layout(height=900, width=900)
fig.show()
```

### **Matriz Correlacion (Variables Seleccionadas)**

```python
selected_vars = [
    "MemoryComplaints",
    "BehavioralProblems",
    "SleepQuality",
    "MMSE",
    "ADL",
    "FunctionalAssessment",
    "Age",
    "BMI",
    "AlcoholConsumption",
    "Diagnosis"
]

# Matriz de correlación
corr_reduced = data[selected_vars].corr(numeric_only=True)

# Máscara para ocultar la mitad superior
mask = np.triu(np.ones_like(corr_reduced, dtype=bool))
corr_half = corr_reduced.mask(mask)

# Crear figura
plt.figure(figsize=(10, 8))

# Dibujar heatmap
plt.imshow(corr_half, aspect='auto')

# Etiquetas de los ejes
plt.xticks(range(len(selected_vars)), selected_vars, rotation=45, ha="right")
plt.yticks(range(len(selected_vars)), selected_vars)

# Barra de color
plt.colorbar()
# Guardar imagen
plt.savefig("matriz_correlacion_diagnosis.png", dpi=300, bbox_inches="tight")

# Mostrar
plt.show()
```

**Conclusión**

Analizando la correlación de cada variable con la variable objetivo Diagnosis (Alzheimer):

**1. Variables con correlación positiva moderada:**

MemoryComplaints (0.31) y BehavioralProblems (0.22)
→ Los pacientes que presentan quejas de memoria y problemas de comportamiento tienden a tener un mayor riesgo de Alzheimer.

**2. Variables con correlación negativa moderada:**

MMSE (-0.24), ADL (-0.33) y FunctionalAssessment (-0.36)
→ Las puntuaciones cognitivas y funcionales más altas se asocian con menos riesgo de Alzheimer.
→ Esto es coherente: a menor deterioro cognitivo y funcional, menor probabilidad de diagnóstico.

**3. Variables con correlación débil:**

La mayoría de los factores clínicos y demográficos (BMI, Hypertension, Diabetes, Edad, Género, Colesterol, etc.) presentan correlaciones muy bajas (cercanas a 0).
→ Esto indica que por sí solas no son buenos predictores de Alzheimer, pero podrían ser útiles combinadas en un modelo multivariado.

**4. Variables de estilo de vida y antecedentes familiares:**

FamilyHistoryAlzheimers (-0.03), Smoking (-0.004), AlcoholConsumption (-0.008)
→ Muy baja correlación individual, por lo que no son determinantes por sí mismas en este dataset.

### **Relación entre Genero y Diagnostico**

---



```python
data['Diagnosis_Label'] = data['Diagnosis'].map({0: 'No Alzheimer', 1: '  Yes Alzheimer'})
# También creamos etiquetas legibles para Gender
data['Gender_Label'] = data['Gender'].map({0: 'Male', 1: 'Female'})

# --- PASO 2: Generar el Gráfico ---
plt.figure(figsize=(7, 6))

# Usamos la nueva columna Gender_Label en el eje X y Diagnosis_Label para el color (hue)
sns.countplot(
    data=data,
    x="Gender_Label",
    hue="Diagnosis_Label",
    palette="coolwarm" # Un esquema de color atractivo
)

plt.title("Relación entre Género y Diagnóstico de Alzheimer", fontsize=14)
plt.xlabel("Gender", fontsize=12)
plt.ylabel("Conteo de Pacientes", fontsize=12)
plt.legend(title="Alzheimer's diagnosis")
plt.grid(axis='y', linestyle='--', alpha=0.7) # Añadir una cuadrícula suave
plt.tight_layout() # Ajustar el diseño para evitar solapamiento
plt.show()
```

**Conclusión**

*  La muestra tiene un número similar de hombres (650) y mujeres (700).

*  En ambos géneros, la mayoría de los pacientes no tiene Alzheimer, mientras que los casos diagnosticados son menores.

*  La proporción de diagnóstico de Alzheimer es muy similar entre hombres y mujeres, por lo que el género no parece ser un factor determinante en este dataset.

*  En números absolutos, los casos de Alzheimer son ligeramente más altos en hombres (380) que en mujeres (370), pero la diferencia es mínima.

Es decir, el riesgo de Alzheimer en este dataset es comparable entre hombres y mujeres.
